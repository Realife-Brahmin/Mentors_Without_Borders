# Exercise 3-2
Write a function named rightjustify that takes a string named s as a parameter and prints the string with enough leading spaces so that the last letter of the string is in column 70 of the display.

```julia
function rightjustify(string)
    println(" "^(70-length(string))*string)
end
rightjustify("8th January 2023")
```

```julia
function firstLine()
    println("+", "-"^4,"+","-"^4,"+")
end
function secondLine()
    println("|", " "^4,"|"," "^4,"|")
end

function printgrid()
    firstLine()
    secondLine()
    secondLine()   
    secondLine()
    secondLine()       
    firstLine()
    secondLine()
    secondLine()   
    secondLine()
    secondLine() 
    firstLine()

end
printgrid()
```
