# Exercice 10 - 1

```julia
function nestedsum(arr)
    total_sum = 0
    for intern_arr in arr
        for item in intern_arr
            total_sum += item
        end
    end
    return total_sum
end
t = [[1, 2], [3], [4, 5, 6]];
nestedsum(t)
```

# Exercise 10-2


```julia
function cumulsum(arr)
    newArr = arr[:]
    sum = 0
    for i in eachindex(arr)
        sum += arr[i]
        newArr[i]= sum    
    end
    return newArr
end

t = [1, 2, 3]
print(cumulsum(t))
```

# Exercise 10-5

```julia
function issort(arr)
    for i in 1:length(arr)-1
        if arr[i] > arr[i+1]
            return false
        end
    end
    return true
end
print(issort([1, 2, 2]))
print(issort(['a', 'd', 'c']))
```

# Exercise 10-6

```julia
function isanagram(str1, str2)
    arr1 = sort(collect(str1))
    arr2 = sort(collect(str2))
    return arr1 == arr2
end
```

# Exercise 10-7

```julia
function hasduplicates(arr)
    unique_elements = Set(arr)
    return length(unique_elements) != length(arr)
end
```

# Exercise 10-8

```julia
function birthday_paradox(n;sample=1000)
    count_matches = 0
    number_students = n
    for _ in 1:sample
        birthdays = rand(1:365, number_students)
        unique_birthdays = Set(birthdays)
        
        if length(birthdays) != length(unique_birthdays)
            count_matches += 1
        end
    end
    
    probability = count_matches / sample
    return probability
end

chance = birthday_paradox(70,200)
println("Chances that two of you have the same birthday is : $chance")

```
